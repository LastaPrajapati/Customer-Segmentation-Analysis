```python
# Ensure necessary features are selected
customer_features = df[['Income']]  # Select 'Income' as a feature

# Create 'In_relationship' as a binary feature
customer_features['In_relationship'] = df[['marital_Together', 'marital_Married']].sum(axis=1)

# Create the 'MntTotal' feature as the sum of all spending columns
df['MntTotal'] = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

# Add 'MntTotal' to the selected features in customer_features
customer_features = customer_features.join(df['MntTotal'])

# Display the first few rows to verify
print(customer_features.head())
```

Feature Scaling, Determining Optimal Number of Clusters and Applying K-means Clustering


```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(customer_features)

# Determine the optimal number of clusters using the Elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Apply K-means with the optimal number of clusters (4)
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
clusters = kmeans.fit_predict(scaled_features)

# Add cluster labels to the dataframe
customer_features['Cluster'] = clusters
```

Analyzing the characteristics of each cluster

```python

cluster_summary = customer_features.groupby('Cluster').mean()
cluster_size = customer_features['Cluster'].value_counts(normalize=True) * 100

print(cluster_summary)
print(cluster_size)
```

```python
# Summary of each cluster based on the analysis
cluster_summary['Cluster Size (%)'] = cluster_size

# Display cluster characteristics
for cluster_id in range(4):
    print(f"Cluster {cluster_id}:")
    print(f"- Average Income: {cluster_summary.loc[cluster_id, 'Income']}")
    print(f"- In_relationship: {cluster_summary.loc[cluster_id, 'In_relationship']}")
    print(f"- Average Total Spending (MntTotal): {cluster_summary.loc[cluster_id, 'MntTotal']}")
    print(f"- Percentage of Customers: {cluster_summary.loc[cluster_id, 'Cluster Size (%)']:.2f}%\n")

# Visualization
sns.scatterplot(x='MntTotal', y='Income', hue='Cluster', data=customer_features, palette='viridis')
plt.title('Customer Segments based on Total Spending and Income')
plt.xlabel('Total Spending (MntTotal)')
plt.ylabel('Income')
plt.show()
```
